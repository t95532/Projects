```python
def outliers_(dataframe, column):
    Q1 = np.percentile(dataframe[column], 25)
    Q3 = np.percentile(dataframe[column], 75)
    """calculating the IQR"""
    IQR = Q3 - Q1
    '''Define lower and upper bounds'''
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    """return outliers in the data & count of outliers"""
    outliers = (dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)
    return outliers
```

```python
outliers = outliers_(data, 'Salary')
```

```python
#No of Outliers
print('Number. of outliers :', len(data['Salary'][outliers].index))
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(data['Salary'][~outliers].index, data['Salary'][~outliers], color='blue', label='Non-Outliers')
plt.scatter(data['Salary'][outliers].index, data['Salary'][outliers], color='red', label='Outliers')
plt.xlabel('index')
plt.ylabel('Salary')
plt.title('Scatter Plot with Outliers')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import datetime as dt
```

```python
data['DOB'] = pd.to_datetime(data['DOB'], format='mixed')

ages = [round((pd.to_datetime(dt.date.today())-data['DOB'][i]).days/365.25) for i in range(0, len(data))]

data['Ages'] = ages
```

```python
plt.figure(figsize=(10,6))
sns.displot(data['Ages'], kde=True)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='ID', y='Ages', hue='Gender')
```

```python
data['DOL'] = data['DOL'].replace('present', dt.date.today())
```

```python
data['DOL'] = pd.to_datetime(data['DOL'], format = 'mixed')
```

```python
data['DOJ'] = pd.to_datetime(data['DOJ'], format = 'mixed')
```

```python
data['Experience']= [np.abs(round((data['DOL'][i]-data['DOJ'][i]).days/365 ,1)) for i in range(0, len(data))]
```
