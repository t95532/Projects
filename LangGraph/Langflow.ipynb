```python
workflow = Graph()
```

```python
# we pass input to function1
workflow.add_node(function1)
# wahtever output of function1 goes as input to function2
workflow.add_node(function2)
#edge represents relationship between nodes
workflow.add_edge("llm","upper_string")
#workflow.add_edge("upper_string","llm")
```

```python
workflow.set_entry_point("llm")
```

```python
workflow.set_finish_point("upper_string")
```

```python
#compile and run the workflow
app = workflow.compile()
```
